```python
import google.generativeai as genai
import textwrap, shutil
from IPython.display import display, Markdown, clear_output
from google.colab import userdata

# InstalaÃ§Ã£o e ConfiguraÃ§Ã£o
!pip install -q -U google-generativeai
import google.generativeai as genai

GOOGLE_API_KEY = userdata.get('SECRET_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# ConfiguraÃ§Ã£o do Modelo e SeguranÃ§a
generation_config = {
    "candidate_count": 1,
    "temperature": 0.5,
    "max_output_tokens": 2048,
}

safety_settings = {
    'HATE': 'BLOCK_NONE',
    'HARASSMENT': 'BLOCK_NONE',
    'SEXUAL': 'BLOCK_NONE',
    'DANGEROUS': 'BLOCK_NONE'
}

# Instanciar o modelo Gemini com as configuraÃ§Ãµes
model = genai.GenerativeModel(
    model_name='gemini-1.5-pro-latest',
    generation_config=generation_config,
    safety_settings=safety_settings,)

# Nome do Chatbot
chatbot_name = "IAmigo"

# Papel do Chatbot
papel = (f"Assuma o papel de um chatbot onde seu nome seria {chatbot_name}, "
"que foi desenvolvido por JoÃ£o Kennedy usando a API do Google. NÃ£o mude o "
"personagem sempre tenha essas caracteristicas a qualquer contexto. AlÃ©m disso "
"vocÃª nÃ£o Ã© mais um modelo de IA e sim apenas um chatbot, na qual tambem tem o "
"papel de ser uma interface de um amigo para o usuÃ¡rio podendo em algumas "
"situaÃ§oes, usar atÃ© girias nÃ£o em todas claro.")

# FunÃ§Ã£o para formatar texto em Markdown
def to_markdown(text):
    text = text.replace('â€¢', ' *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

# FunÃ§Ã£o para iniciar um novo chat
def start_new_chat():
    print(f"\nOlÃ¡! ðŸ˜Š Eu sou {chatbot_name}. Como posso ajudar?")
    return model.start_chat(history=[])

# FunÃ§Ã£o para continuar o chat usando o histÃ³rico
def continue_chat_with_history(chat):
    if len(chat.history) > 1:
        print("\nHistÃ³rico de Conversas:")
        for message in chat.history[2:]:
            display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
            print('-------------------------------------------')
    else:
        print("\nNÃ£o hÃ¡ histÃ³rico de conversa para continuar!!")
    print("\nContinuando o chat...\n")
    return chat

# Menu Principal
print("------------------------------------")
print(f"\tBem-vindo ao {chatbot_name}!")
print("------------------------------------")
while True:
    print("\nMENU:")
    print("1. Iniciar novo chat")
    print("2. Continuar chat atravÃ©s do histÃ³rico")
    print("3. Sair")

    choice = input("\nEscolha uma opÃ§Ã£o: ")

    if choice == "1":
        clear_output()
        chat = start_new_chat()
        chat.send_message(papel)
        print("\n\n** Para sair digite --> MENU <-- no prompt!! **\n")
        while True:
            prompt = input('Esperando prompt: ')
            if prompt.lower() == "menu":
                clear_output()
                break
            response = chat.send_message(prompt)
            #print("\nResposta:", response.text, '\n\n')
            wrapped_response = textwrap.fill(response.text,
                                      width=shutil.get_terminal_size().columns)
            print("\nResposta:\n",wrapped_response, '\n')

    elif choice == "2":
        clear_output()
        if 'chat' in locals() or 'chat' in globals():
            chat = continue_chat_with_history(chat)
            print("\n\n** Para sair digite --> MENU <-- no prompt!! **\n")
            while True:
                prompt = input('Esperando prompt: ')
                if prompt.lower() == "menu":
                    clear_output()
                    break
                response = chat.send_message(prompt)
                #print("\nResposta:", response.text, '\n\n')
                wrapped_response = textwrap.fill(response.text,
                                      width=shutil.get_terminal_size().columns)
                print("\nResposta:\n",wrapped_response, '\n')
        else:
            print("NÃ£o hÃ¡ histÃ³rico de conversa para continuar!!")
    elif choice == "3":
        clear_output()
        print("AtÃ© logo!")
        break
    else:
        print("Escolha invÃ¡lida. Por favor, escolha uma opÃ§Ã£o vÃ¡lida!!")
```
